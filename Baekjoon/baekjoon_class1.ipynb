# 2675 문자열 반복

```python
t = int(input())

for n in range(t):
    r, s = input().split()
    slist = []
    for i in range(len(list(s))):    
        for j in range(int(r)):
            slist.append(list(s)[i])
    
    print(''.join(slist))
```

```python
test = int(input())

for i in range(test):
    li = []
    mul, insert = input().split()
    mul = int(mul)
    for s in insert:
        li.append(s*mul)
    
    result = ''.join(li)
    print(result)
```

# 1157 단어공부

```python
str1 = input().upper()
dict1 = {}

for i in range(len(str1)):
    num1 = str1.count(str1[i])
    dict1[str1[i]] = int(num1)

list1 = list(dict1.values())
set1 = set(list1)

if len(list1) > len(set1):
    print('?')
else :
    print(max(dict1, key=dict1.get))
```

```python
str1 = input().upper()
str_list = []
num_list = []

for i in range(len(str1)):
    if str1[i] not in str_list:
        num1 = str1.count(str1[i])
        num_list.append(num1)
        str_list.append(str1[i])

my_max = max(num_list)

if num_list.count(my_max) > 1 :
    print('?')
else :
    idx = num_list.index(my_max)
    print(str_list[idx])
```

## 10809 알파벳 찾기
